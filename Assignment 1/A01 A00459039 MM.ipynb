```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import csv
```

Read an image file with a colour-depth of at least 24-bit.
Display the image and its shape

```python
image = img.imread(inpath+"c_small.png")
plt.axis('off')
plt.imshow(image)
plt.show()
image.shape
```

Extract RGB colour components:

```python
allcol = image.reshape(-1,3)
print(allcol.shape)
print(allcol)
```

Find set of unique colours:

```python
allcol_unique, allcol_index = np.unique(allcol, return_counts=True, axis=0)
print(allcol_unique.shape)
print(allcol_unique)
```

Find frequency of unique colours and display line graph:

```python
print(allcol_index)
```

```python
au = np.zeros((75625,3))
au[:75533] = allcol_unique
print(au.shape)
print(au)
```

```python
au_reshaped = au.reshape(275, 275, 3)
plt.imshow(au_reshaped, vmin=0, vmax=255)
```

Export data to 'all_colours.csv', 'unique_colours.csv', and 'unique_freq.csv'

```python
f = open(outpath+"all_colours.csv", 'w')
w = csv.writer(f)
w.writerows(allcol)
f.close()

f = open(outpath+"unique_colours.csv", 'w')
w = csv.writer(f)
w.writerows(allcol_unique)
f.close()

f = open(outpath+"unique_freq.csv", 'w')
w = csv.writer(f)
w.writerow(allcol_index)
f.close()
```

Reduce original image colour depth to 3-bit for each component, then repeat above steps

```python
img_small = (image)*7
img_small2 = np.array(img_small, int)
img_small2 = img_small2/img_small2.max()
plt.imshow(img_small2)
```

```python
sml_col = img_small2.reshape(-1,3)
print(sml_col.shape)
print(sml_col)
```

```python
smlcol_unique, smlcol_index = np.unique(sml_col, return_counts=True, axis=0)
print(smlcol_unique.shape)
print(smlcol_unique)
```

```python
au = np.zeros((81,3))
au[:66] = smlcol_unique
print(au.shape)
print(au)
```

```python
au_reshaped = au.reshape(9, 9, 3)
plt.imshow(au_reshaped)
```

```python
f = open(outpath+"all_colours_3bit.csv", 'w')
w = csv.writer(f)
w.writerows(sml_col)
f.close()

f = open(outpath+"unique_colours_3bit.csv", 'w')
w = csv.writer(f)
w.writerows(smlcol_unique)
f.close()

f = open(outpath+"unique_freq_3bit.csv", 'w')
w = csv.writer(f)
w.writerow(smlcol_index)
f.close()
```
