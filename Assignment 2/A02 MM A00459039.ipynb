Write Python code to generate a video of frames with a filled circle that changes its radius from 10px to the video width (or height, whichever the smaller)

```python
# Define video parameters
width = 300
height = 400
fps = 20
# Empty list to hold frames
images_list = []
# First frame bg (video dimensions, (0,0,0) colour)
image = np.zeros((height,width,3))
# Initialize center coords and radius of circle
xpos, ypos, rad = 10,10,10

while (xpos+rad) <= 300:
  # Create new frame with red circle on frame bg
  newimg = cv2.circle(image, (xpos,ypos), rad, (255,0,0), -1)
  # Append new frame to video
  images_list.append(newimg)
  # Reset frame bg to pure black for next frame
  image = np.zeros((height,width,3))
  # Increment center coords and radius so that circle grows and stays in-frame
  xpos += 1
  ypos += 1
  rad += 1

# Generate clip
clip = mpy.ImageSequenceClip(images_list, fps=20)
mpy.ipython_display(clip)
```

Write Python code that reads an existing video and overlays a filled-circle that moves.

```python
# Import video, extract parameters
vid = mpy.VideoFileClip(inpath+"sky.mp4")
v_width = vid.size[0]
v_height = vid.size[1]
v_fps = vid.fps
v_dur = vid.duration # <- seconds, float
v_frames = vid.fps*vid.duration
```

```python
# Define circle position, movement, radius
xpos, ypos = 21, 21
xdir, ydir = 5, 5
rad = 20
# Empty list to hold frames for new vid
images_list = []

for index in range (0,int(v_frames)):
  # Get next frame
  image = vid.get_frame(index/v_fps)
  # Add circle to frame
  circled = cv2.circle(image, (xpos,ypos), rad, (255,150,0), -1)
  # Append frame with circle to list of frames for new vid
  images_list.append(circled)
  # Reverse direction if edge hit
  if (xpos-rad) <= 0 or (xpos+rad) >= v_width:
    xdir = -xdir
  if (ypos-rad) <= 0 or (ypos+rad) >= v_height:
    ydir = -ydir

  # Increment position by direction
  xpos += xdir
  ypos += ydir

# Generate new clip
clip = mpy.ImageSequenceClip(images_list, fps=v_fps)
mpy.ipython_display(clip)
```

Write Python code that reads a video and overlays a mask layer with a "cut-out" circle, portions of the original video within which (the circle) are only shown (and the rest is black/masked)

```python
# Import video, extract parameters
vid = mpy.VideoFileClip(inpath+"sky.mp4")
v_width = vid.size[0]
v_height = vid.size[1]
v_fps = vid.fps
v_dur = vid.duration # <- seconds, float
v_frames = vid.fps*vid.duration
```

```python
# Define circle position, movement, radius
xpos, ypos = 21, 21
xdir, ydir = 5, 5
rad = 20
# Initialize mask
mask = np.zeros((v_height,v_width), dtype="uint8")
cv2.circle(mask, (xpos,ypos), rad, (255,255,255), -1)
# Empty list to hold frames for new vid
images_list = []

for index in range (0,int(v_frames)):
  # Get next frame
  image = vid.get_frame(index/v_fps)
  # Add mask to frame
  cv2.circle(mask, (xpos,ypos), rad, (255,255,255), -1)
  masked = cv2.bitwise_and(image, image, mask=mask)
  # Add frame to new video
  images_list.append(masked)
  # Re-initialize mask
  mask = np.zeros((v_height,v_width), dtype="uint8")
  # Reverse circle direction if edge hit
  if (xpos-rad) <= 0 or (xpos+rad) >= v_width:
    xdir = -xdir
  if (ypos-rad) <= 0 or (ypos+rad) >= v_height:
    ydir = -ydir
  # Increment circle position
  xpos += xdir
  ypos += ydir

# Generate new video
clip = mpy.ImageSequenceClip(images_list, fps=v_fps)
mpy.ipython_display(clip)
```
